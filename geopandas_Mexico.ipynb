```python
import pandas as pd
import geopandas as gpd 
```

```python
#Read in shapefile
#data source : https://gadm.org/download_country.html 
Mexican_states=gpd.read_file('gadm36_MEX_1.shp')
Mexican_states
```

```python
type(Mexican_states)
```

```python
Mexican_states=Mexican_states.drop(['GID_0', 'NAME_0', 'GID_1', 'VARNAME_1', 'NL_NAME_1', 'TYPE_1', 'ENGTYPE_1', 'CC_1', 'HASC_1'],
                                  axis=1)
```

```python
Mexican_states
```

```python
Mexican_states.rename(columns={'NAME_1': 'State'}, inplace=True)
```

```python
Mexican_states.insert(0, 'entidad', range(1, 1 + len(Mexican_states)))
```

```python
Mexican_states
```

```python
My_Mexican_dataset=pd.read_csv('my_Mexican_dataset.csv')
My_Mexican_dataset
```

```python
Merged_dfs = pd.merge(left=My_Mexican_dataset, right=Mexican_states, left_on='entidad', right_on='entidad')

Merged_dfs
```

```python
#Merged_dfs=Merged_dfs.drop(['State_x', 'Female2010', 'Female2020', 'Male2010', 'Male2020'], axis=1)
Merged_dfs = Merged_dfs.loc[Merged_dfs.index.drop_duplicates()]
```

```python
type(Merged_dfs)
```

```python
Mexico_df=Merged_dfs
```

```python
Mexico_df
```

```python
from geopandas import GeoDataFrame
from shapely.geometry import Point

geometry = Mexico_df['geometry']

Mexico_gdf = GeoDataFrame(Mexico_df, crs="EPSG:4326", geometry=geometry)
```

```python
Mexico_gdf
```

```python
#Population 2010
#Mexico_gdf.plot(column='Pop2010', cmap='twilight', figsize=(8,8), legend=True)

# Resizing the legend 
import matplotlib.pyplot as plt 
from mpl_toolkits.axes_grid1 import make_axes_locatable 

fig, ax = plt.subplots(figsize = (10,10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size = "7%", pad = 0.1)
Mexico_gdf.plot(column='Pop2010', cmap='twilight', legend=True, ax = ax, cax = cax)

plt.savefig('Population_2010.png', dpi=200)
```
