<h4> To evaluate whether this difference is statistically significant <h4>

# t -test
- Question 1: First, we want to know whether in the total population of Mexico the change between the number of people who spoke an indigenous language in 2010 compared to 2020 is significant.

H0 : µ2010 = µ2020

Ha : µ2010 =! µ2020

In order to test our null hypothesis, we first need to stablish our level of confidence. In this case, I, as an analyst, choose a confidence level of 95 %, that is, I give my experiment an error level as a tolerance of 5 %.

```python
import scipy.stats as st
```

```python
#We test our null hypothesis by using a t-test.
#"Probability of getting a difference between sample means is",
#Prob = 1 - st.t.cdf(t, ddl)) 
# to calculate t:

XLang2010 = Merged_Pops_lang['Lang_prmillion2010'].mean()
XLang2020= Merged_Pops_lang['Lang_prmillion2020'].mean()
SLang2010 = Merged_Pops_lang['Lang_prmillion2010'].std()
SLang2020= Merged_Pops_lang['Lang_prmillion2020'].std()

print("the mean XLang2010 = %f and XLang2020 = %f" % (XLang2010, XLang2020))
print("the std  SLang2010 = %f and SLang2020 = %f" % (SLang2010 , SLang2020))
      
n1=32
n2=32
dof=n1+n2-2
ccdof= np.sqrt((1/n1) +(1/n2))
Spooled= np.sqrt(((n1-1)*((SLang2010)**2) +(n2-1)*((SLang2020)**2))/((n1+n2)-2))
et=Spooled*ccdof

t = (XLang2010 - XLang2020)/et
print ("the t value is t = %f" % t)

#since the total degrees of freedom are dof=(n1+n2)-2.  

Prob_gauche=st.t.cdf(t,dof )

Prob_droit=1-Prob_gauche

print("Prob=%f and (1-Prob)=%f "% (Prob_gauche, Prob_droit))   

```

```python
#statistic_two-sided, pvalue_two-sided = st.ttest_ind(df['pre.weight'], df['weight6weeks'], alternative='two-sided')
statistic_greater, pvalue_greater= st.ttest_ind(Merged_Pops_lang['Lang_prmillion2010'], Merged_Pops_lang['Lang_prmillion2020'], alternative = 'greater')
statistic_less, pvalue_less = st.ttest_ind(Merged_Pops_lang['Lang_prmillion2010'], Merged_Pops_lang['Lang_prmillion2020'], alternative = 'less')

print(statistic_greater, pvalue_greater)
print(statistic_less, pvalue_less)
```

Since p value > 0.05, we conclude that, with 95 percent confidence, there has not been a significant change in the number of people who speak an indigenous language between 2010 and 2020. In other words, we can not reject our null hypothesis.
